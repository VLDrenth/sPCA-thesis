#### `SPA Tests`, `MCS`, and `Treatment effects`


```python
import numpy as np
import pandas as pd
```

```python
# Load errors from results
targets = ['inflation', 'unemployment', 'ip_growth']
horizons = [1, 6, 12]
scaling = ['regression', 'distance_correlation']
dim_methods = ['pca', 'rbf', 'sigmoid', 'ae']
forecast_methods = ['ols']

mse_df = pd.DataFrame(columns=['target', 'horizon', 'scale_method', 'dim_method', 'forecast_method', 'MSE'])

for target in targets:
    for horizon in horizons:
        for scale_method in scaling:
            for dim_method in dim_methods:
                for forecast_method in forecast_methods:
                    try:
                        errors_path = 'resources/results/errors/{}_{}_{}_{}_h{}.npy'.format(target, dim_method, scale_method, forecast_method, horizon)
                        errors = np.load(errors_path)
                    except:
                        print('Error loading: {}'.format(errors_path))
                        continue

                    mse = np.mean(errors**2)

                    mse_df = mse_df.append({'target': target, 'horizon': horizon, 'scale_method': scale_method, 'dim_method': dim_method,
                                            'forecast_method': forecast_method, 'MSE': mse}, ignore_index=True)
```

```python
pd.set_option('display.float_format', lambda x: '%.10f' % x)
print(mse_df.to_string())
```
